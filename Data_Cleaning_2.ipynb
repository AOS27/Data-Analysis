```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"

headers = ['Manufacturer', 'Category', 'Screen', 'GPU', 'OS', 'CPU_core', 'Screen_Size_cm', 'CPU_frequency', 'RAM_GB', 'Storage_GB_SSD', 'Weight_kg', 'Price'
]

df = pd.read_csv(filepath, names=headers, header=0)
```

```python
print(df.info())
```

```python
df.head()
```

```python
df['Screen_Size_cm']=np.round(df['Screen_Size_cm'], 2)
df.head()
```

```python
missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
df[['Weight_kg','Screen_Size_cm']] = df[['Weight_kg','Screen_Size_cm']].astype('float')
```

```python
df.dtypes
```

```python
df['Weight_kg'] = df['Weight_kg']*2.205
df['Screen_Size_cm'] = df['Screen_Size_cm']/2.54
df.rename(columns={'Weight_kg': 'Weight_pounds', 'Screen_Size_cm': 'Screen_Size_In'}, inplace=True)
```

```python
df.head(50)
```

```python
weight_avg = df['Weight_pounds'].mean(axis=0)
weight_avg
```

```python
df['Weight_pounds'].replace(np.nan, weight_avg, inplace=True)
```

```python
Screen_Size_freq = df['Screen_Size_In'].value_counts().idxmax()
Screen_Size_freq

df['Screen_Size_In'].replace(np.nan, Screen_Size_freq, inplace=True)
```

```python
df.head(50)
```

```python
df[['Screen_Size_In','Weight_pounds']] = df[['Screen_Size_In','Weight_pounds']].round(2)
```

```python
df.head()
```

```python
df['CPU_frequency'] = df['CPU_frequency']/df['CPU_frequency'].max()
```

```python
bins = np.linspace(min(df['Price']), max(df['Price']), 4)
group_names = ['Low', 'Medium', 'High']

df['Price-binned']=pd.cut(df['Price'], bins, labels=group_names, include_lowest=True)
df[['Price', 'Price-binned']].head(50)
```

```python
plt.bar(group_names, df["Price-binned"].value_counts())
plt.xlabel("Price")
plt.ylabel("count")
plt.title("Price bins")
```

```python
dummies = pd.get_dummies(df['Screen'])
df = pd.concat([df, dummies], axis=1)
df.rename(columns={'Full HD':'Screen-Full_HD', 'IPS Panel':'Screen-IPS_panel'}, inplace=True)
df.drop('Screen', axis = 1, inplace=True)
```
